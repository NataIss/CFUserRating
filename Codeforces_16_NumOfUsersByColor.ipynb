```python
import json
import requests
import pandas as pd
import numpy as np
import time
```

```python
try:  
    data = pd.read_json("https://codeforces.com/api/contest.list?gym=false")
    con = pd.DataFrame.from_records(data['result'])
    con
except (BaseException) as e:    
    print(e)
con
```

```python
con['d']=pd.to_datetime(con['startTimeSeconds'],unit='s')
```

```python
con=con[(con['d']<='2021-12-31') & (con['d']>'2018-07-03')]
con.shape
```

```python
 link='https://codeforces.com/api/contest.ratingChanges?contestId=1601'
 data = pd.read_json(link)
 if data['status'][0]=='OK':
       dfAll = pd.DataFrame.from_records(data['result'])
```

```python
for j in con['id']: 
    i=0
    while i<3:
        try:
            time.sleep(0.2)
            link='https://codeforces.com/api/contest.ratingChanges?contestId=' + str(j)
            data = pd.read_json(link)
            if data['status'][0]=='OK':
                df = pd.DataFrame.from_records(data['result'])
                dfAll=pd.concat([dfAll,df],ignore_index=True)
                i=i+5
        except (BaseException) as e:    
                time.sleep(1)
                i=i+1
                #ratedListAll[ratedListAll['handle']==j][0]
                if i==3:
                    print(j)
                    print(e)
dfAll
```

```python
dfAll.drop_duplicates(keep='last',subset=['handle','contestId'],inplace=True)
dfAll['d']=pd.to_datetime(dfAll['ratingUpdateTimeSeconds'],unit='s')
dfAll
```

```python
dfAll['title']='newbie'
dfAll['color']='gray'
dfAll['title'].where(~(dfAll.newRating >= 3000), "legendary grandmaster", inplace=True)
dfAll['title'].where(~((dfAll.newRating >= 2600)&(dfAll.newRating < 3000)), "international grandmaster", inplace=True)
dfAll['title'].where(~((dfAll.newRating >= 2400)&(dfAll.newRating < 2600)), "grandmaster", inplace=True)
dfAll['title'].where(~((dfAll.newRating >= 2300)&(dfAll.newRating < 2400)), "international master", inplace=True)
dfAll['title'].where(~((dfAll.newRating >= 2100)&(dfAll.newRating < 2300)), "master", inplace=True)
dfAll['title'].where(~((dfAll.newRating >= 1900)&(dfAll.newRating < 2100)), "candidate master", inplace=True)
dfAll['title'].where(~((dfAll.newRating >= 1600)&(dfAll.newRating < 1900)), "expert", inplace=True)
dfAll['title'].where(~((dfAll.newRating >= 1400)&(dfAll.newRating < 1600)), "specialist", inplace=True)
dfAll['title'].where(~((dfAll.newRating >= 1200)&(dfAll.newRating < 1400)), "pupil", inplace=True)
dfAll['color'].where(~(dfAll.newRating >= 3000), "red", inplace=True)
dfAll['color'].where(~((dfAll.newRating >= 2600)&(dfAll.newRating < 3000)), "red", inplace=True)
dfAll['color'].where(~((dfAll.newRating >= 2400)&(dfAll.newRating < 2600)), "red", inplace=True)
dfAll['color'].where(~((dfAll.newRating >= 2300)&(dfAll.newRating < 2400)), "orange", inplace=True)
dfAll['color'].where(~((dfAll.newRating >= 2100)&(dfAll.newRating < 2300)), "orange", inplace=True)
dfAll['color'].where(~((dfAll.newRating >= 1900)&(dfAll.newRating < 2100)), "violet", inplace=True)
dfAll['color'].where(~((dfAll.newRating >= 1600)&(dfAll.newRating < 1900)), "blue", inplace=True)
dfAll['color'].where(~((dfAll.newRating >= 1400)&(dfAll.newRating < 1600)), "cyan", inplace=True)
dfAll['color'].where(~((dfAll.newRating >= 1200)&(dfAll.newRating < 1400)), "green", inplace=True)
```

```python
dfAll
```

```python
df18=dfAll[(dfAll['d']<='2018-12-31') & (dfAll['d']>'2018-07-03')]
df18.drop_duplicates(keep='first',subset=['handle'],inplace=True)
res18=pd.DataFrame(df18.groupby('title')['handle'].count())
res18.rename(columns={'handle':'2018'}, inplace=True)
res18
```

```python
col18=pd.DataFrame(df18.groupby('color')['handle'].count())
col18.rename(columns={'handle':'2018'}, inplace=True)
col18
```

```python
df19=dfAll[(dfAll['d']<='2019-12-31') & (dfAll['d']>'2019-07-03')]
df19.drop_duplicates(keep='first',subset=['handle'],inplace=True)
res19=pd.DataFrame(df19.groupby('title')['handle'].count())
res19.rename(columns={'handle':'2019'}, inplace=True)
col19=pd.DataFrame(df19.groupby('color')['handle'].count())
col19.rename(columns={'handle':'2019'}, inplace=True)
```

```python
df20=dfAll[(dfAll['d']<='2020-12-31') & (dfAll['d']>'2020-07-03')]
df20.drop_duplicates(keep='first',subset=['handle'],inplace=True)
res20=pd.DataFrame(df20.groupby('title')['handle'].count())
res20.rename(columns={'handle':'2020'}, inplace=True)
col20=pd.DataFrame(df20.groupby('color')['handle'].count())
col20.rename(columns={'handle':'2020'}, inplace=True)
```

```python
df21=dfAll[(dfAll['d']<='2021-12-31') & (dfAll['d']>'2021-07-03')]
df21.drop_duplicates(keep='first',subset=['handle'],inplace=True)
res21=pd.DataFrame(df21.groupby('title')['handle'].count())
res21.rename(columns={'handle':'2021'}, inplace=True)
col21=pd.DataFrame(df21.groupby('color')['handle'].count())
col21.rename(columns={'handle':'2021'}, inplace=True)
```

```python
res=pd.merge(res18,res19,how="inner",left_index=True,right_index=True)
res=pd.merge(res,res20,how="inner",left_index=True,right_index=True)
res=pd.merge(res,res21,how="inner",left_index=True,right_index=True)
res
```

```python
col=pd.merge(col18,col19,how="inner",left_index=True,right_index=True)
col=pd.merge(col,col20,how="inner",left_index=True,right_index=True)
col=pd.merge(col,col21,how="inner",left_index=True,right_index=True)
col
```

```python
res.to_csv('res.csv')
col.to_csv('col.csv')
```
