```python
import json
import requests
import pandas as pd
import numpy as np
import time
```

```python
try:  
    data = pd.read_json("https://codeforces.com/api/contest.list?gym=false")
    con = pd.DataFrame.from_records(data['result'])
    con
except (BaseException) as e:    
    print(e)
con
```

```python
con['d']=pd.to_datetime(con['startTimeSeconds'],unit='s')
```

```python
dfAll=pd.DataFrame()
```

```python
for j in con['id']: 
    i=0
    while i<3:
        try:
            time.sleep(0.2)
            link='https://codeforces.com/api/contest.ratingChanges?contestId=' + str(j)
            data = pd.read_json(link)
            if data['status'][0]=='OK':
                df = pd.DataFrame.from_records(data['result'])
                dfAll=pd.concat([dfAll,df],ignore_index=True)
                i=i+5
        except (BaseException) as e:    
                time.sleep(1)
                i=i+1
                #ratedListAll[ratedListAll['handle']==j][0]
                if i==3:
                    print(str(j) + ' - not rated')
dfAll
```

```python
dfAll['d']=pd.to_datetime(dfAll['ratingUpdateTimeSeconds'],unit='s')
dfAll['date']= [time.date() for time in dfAll['d']]
dfAll['num']=dfAll.sort_values('d').groupby('handle')['handle'].cumcount().add(1)
```

```python
dfAll['title']='10. newbie'
dfAll['color']='7. gray'
dfAll['title'].where(~(dfAll.newRating >= 3000), "1. legendary grandmaster", inplace=True)
dfAll['title'].where(~((dfAll.newRating >= 2600)&(dfAll.newRating < 3000)), "2. international grandmaster", inplace=True)
dfAll['title'].where(~((dfAll.newRating >= 2400)&(dfAll.newRating < 2600)), "3. grandmaster", inplace=True)
dfAll['title'].where(~((dfAll.newRating >= 2300)&(dfAll.newRating < 2400)), "4. international master", inplace=True)
dfAll['title'].where(~((dfAll.newRating >= 2100)&(dfAll.newRating < 2300)), "5. master", inplace=True)
dfAll['title'].where(~((dfAll.newRating >= 1900)&(dfAll.newRating < 2100)), "6. candidate master", inplace=True)
dfAll['title'].where(~((dfAll.newRating >= 1600)&(dfAll.newRating < 1900)), "7. expert", inplace=True)
dfAll['title'].where(~((dfAll.newRating >= 1400)&(dfAll.newRating < 1600)), "8. specialist", inplace=True)
dfAll['title'].where(~((dfAll.newRating >= 1200)&(dfAll.newRating < 1400)), "9. pupil", inplace=True)
dfAll['color'].where(~(dfAll.newRating >= 3000), "1. red", inplace=True)
dfAll['color'].where(~((dfAll.newRating >= 2600)&(dfAll.newRating < 3000)), "1. red", inplace=True)
dfAll['color'].where(~((dfAll.newRating >= 2400)&(dfAll.newRating < 2600)), "1. red", inplace=True)
dfAll['color'].where(~((dfAll.newRating >= 2300)&(dfAll.newRating < 2400)), "2. orange", inplace=True)
dfAll['color'].where(~((dfAll.newRating >= 2100)&(dfAll.newRating < 2300)), "2. orange", inplace=True)
dfAll['color'].where(~((dfAll.newRating >= 1900)&(dfAll.newRating < 2100)), "3. violet", inplace=True)
dfAll['color'].where(~((dfAll.newRating >= 1600)&(dfAll.newRating < 1900)), "4. blue", inplace=True)
dfAll['color'].where(~((dfAll.newRating >= 1400)&(dfAll.newRating < 1600)), "5. cyan", inplace=True)
dfAll['color'].where(~((dfAll.newRating >= 1200)&(dfAll.newRating < 1400)), "6. green", inplace=True)
dfAll
```

```python
dfAll.to_csv('ratingChangesTill2020-02-04WithBase.csv')
```

```python
ratedCon=pd.unique(dfAll['contestId'])
```

```python
con=con[ (con['d']>'2018-07-03')]
con.drop_duplicates(keep='first',subset=['d'],inplace=True)
con.shape
```

```python
resAll=pd.DataFrame()
colAll=pd.DataFrame()
for i, row in con.iterrows():
  if row['id'] in ratedCon:
    x=row['d'].date()
    y=(x-pd.DateOffset(days=180)).date()
    print(x,y)
    for baseNum in [1,5,10,20,30]:
      dfX=dfAll[(dfAll['date']<=x) & (dfAll['date']>y)& (dfAll['num']>=baseNum)]
      dfX.drop_duplicates(keep='first',subset=['handle'],inplace=True)

      resX=pd.DataFrame(dfX.groupby('title')['handle'].count())
      resX.rename(columns={'handle':'count'}, inplace=True)
      resX['dateX']=x
      resX['baseNum']=baseNum
      resX.reset_index(inplace=True)
      colX=pd.DataFrame(dfX.groupby('color')['handle'].count())
      colX.rename(columns={'handle':'count'}, inplace=True)
      colX['dateX']=x
      colX['baseNum']=baseNum
      colX.reset_index(inplace=True)
      
      resAll=pd.concat([resAll,resX],ignore_index=True)
      colAll=pd.concat([colAll,colX],ignore_index=True)
```
